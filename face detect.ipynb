```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('hosna.jpg')
face= cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
eye= cv2.CascadeClassifier("haarcascade_eye.xml")
img1= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2= face.detectMultiScale(img1)
##img3= eye.detectMultiScale(img1)
x= img2[0][0]
y= img2[0][1]
x1= x+img2[0][2]  
y1= y+img2[0][3]
gray_face = img1[y:y1,x:x1]

img3= eye.detectMultiScale(gray_face,minSize=(1,1), scaleFactor=1.1, minNeighbors=100)
out1= cv2.rectangle(img1,(x,y),(x1,y1),(5,5,540),10)

for (xb,yb,w,h) in img3:
   cv2.rectangle(img1,(xb+x,yb+y),(xb+w+x,yb+h+y),(500,51,54),5)



plt.imshow(out1)
plt.show()
```
